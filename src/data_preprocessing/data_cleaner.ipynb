```python
df2=pd.read_csv('data/processed/clean_data_02.csv')
df2.head()
```

```python
df2=df1.loc[(df1['Label'] == 'BENIGN') |  (df1['Label'] == 'PortScan') | (df1['Label'] == 'DDoS') | (df1['Label'] == 'FTP-Patator') | (df1['Label'] == 'SSH-Patator') | (df1['Label'] == 'DoS Slowloris') | (df1['Label'] == 'DoS Slowhttptest') | (df1['Label'] == 'DoS GoldenEye') ]
```

```python
df2['Label'].unique()
```

```python


# Sample only a fraction of BENIGN samples (e.g., 20%)
benign_df = df2[df2['Label'] == 'BENIGN'].sample(frac=0.025, random_state=42)
ddos_df = df2[df2['Label'] == 'DDos'].sample(frac=0.5, random_state=42)
portscan_df = df2[df2['Label'] == 'PortScan'].sample(frac=0.1, random_state=42)
ftp_df = df2[df2['Label'] == 'FTP-Patator'].sample(frac=0.5, random_state=42)
ssh_df = df2[df2['Label'] == 'SSH-Patator'].sample(frac=0.5, random_state=42)
doss_df = df2[df2['Label'] == 'DoS Slowhttptest'].sample(frac=0.5, random_state=42)
dosg_df = df2[df2['Label'] == 'DoS GoldenEye'].sample(frac=0.5, random_state=42)



df3 = pd.concat([benign_df,ddos_df,portscan_df,ftp_df,ssh_df,doss_df,dosg_df]).reset_index(drop=True)
```

```python
df3.shape
```

```python
df3.to_csv('data/processed/clean_data.csv', index=False)
```
